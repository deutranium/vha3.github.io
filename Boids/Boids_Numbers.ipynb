```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Boid numbers

```python
boids = {'Boid Boys': 120,
         'Purple Pandemonium': 1,
         'Awesome Group': 1,
         'Your Neighbor Boid': 1,
         'Volatile': 53,
         'Boids By Daylight': 20,
         'Boids will be Boids': 150,
         'I am Boid in class': 90,
         'Boids of Prey': 100,
         'The chirping chaps': 120,
         'Into the Boid': 75,
         'ZATO': 1}
boids = dict(sorted(boids.items(), key=lambda x: x[1], reverse=True))
```

#### Plot

```python
plt.rcParams['figure.figsize'] = [20, 2.5]
plt.boxplot(list(boids.values()), meanline=True, showmeans=True, whis=1., vert=False, widths=2.5,
            boxprops={'linewidth': 3}, whiskerprops={'color':'black', 'linewidth': 3},
            capprops={'linewidth': 3}, medianprops={'color':'black', 'linewidth': 2},
            meanprops={'linewidth': 2}, flierprops={'markerfacecolor':'black',
                                                                     'markeredgecolor': 'black',
                                                                     'markersize': 10})
plt.title('ECE 4760 Boids Statistics', fontsize=24)
plt.xlabel('Number of Boids at 30fps', fontsize=24)
plt.yticks(fontsize=18, visible=False)
plt.xticks(fontsize=18, visible=True)


# X/y limits
plt.ylim([-(5+len(boids)), 6])

counter = 0 
for i in list(boids.keys()):
    plt.annotate(i, xy=(boids[i], -(5+counter)), xytext=(boids[i], -(5+counter)), fontsize=8, alpha=.8)
    plt.plot(boids[i]*numpy.ones(20), numpy.linspace(-(5+counter),-1,20), 'black', alpha=0.1)
    counter += 1

plt.plot(list(boids.values()), -1.5*numpy.ones(len(list(boids.values()))), 'r.', alpha=0.5, markersize=15)

plt.savefig('Boid.png', bbox_inches='tight')
plt.show()
```
