# Short Answer

###### The Molniya Orbit (shown below) is popular for Russian communications satellites that need to have a line of sight to Russian territory. Explain in terms of Kepler's Second Law why this is a good orbit for such an application.

```python
PATH = "/Users/hunteradams/Documents/PhD Semester 5/Sections/"
Image(filename = PATH + "molniya.png", width=600, height=600)
```

Kepler's Second Law states that a spacecraft will sweep equal areas in equal times. This means that the spacecraft will be moving quickly when near the Earth (perihelion), and slowly when far away (aphelion). As a result, the spacecraft spends most of its time above the Northern Hemisphere (Russia) where it has a view of the Russian territory.

###### A black hole is a very massive object that occupies very little space. Using what you know about differential forces and gravity, explain what would happen to a person's body as he/she fell into one of these objects (hint, think about tides).

Tides on the Earth are caused by differential forces. The sides of the Earth that are closer to the Moon are getting tugged harder than those that are farther away. As a result, each slice of Earth that is equidistant from the Moon is getting pulled away from the slices that are farther away, leading to the stretching out of the Earth (bulging on each side).

As a person (e.g. Matthew McConaughey in *Interstellar*) falls into a black hole, the parts of the body that are closer to the black hole are pulled harder than those that are farther away. As a result, the person is stretched out (sphettified).

In class, we learned that refracting telescopes work by bending incident light to a focal plane, as shown below.

As it turns out, gravity also bends light. This effect (called gravitational lensing) could allow us to use a star much like a telescope. Rather than using optics to bend light to a focus, gravity bends light to a focus. 

#### Part 1

**Assuming an aperature equal to the diameter of the Sun, find the diffraction limit for such a gravitational lensing telescope if the objective is to observe visible light (600 nm).**

\begin{align}
\theta &= \frac{\lambda}{D}\\
&= \frac{600 \times 10^{-9}m}{1.4 \times 10^{9}m}\\
&= 4.29\times 10^{-16} radians
\end{align}

#### Part 2

**The nearest star system with exoplanets is Proxima Centauri, which is 4.2 lightyears ($3.974 \times 10^{16}m$). What is the minimum spatial separation between to objects at this distance that one could theoretically expect to resolve with the Sun as a telescope?**

\begin{align}
\alpha &= \frac{s}{r}
\end{align}

Let $\alpha=\theta$, the diffraction limit. And let $r=3.974\times 10^{16}m$, the distance to the Proxima Centauri system.

\begin{align}
s &= r\alpha\\
&= \left(3.974 \times 10^{16}m\right)\left(4.29 \times 10^{-16}radians\right)\\
&= 17meters
\end{align}

#### Part 3

**So, if the exoplanet around Proxima Centauri came into the field of view of the gravitational telescope, we would be able to resolve objects on the surface as near to one another as the number found in the previous section.**

**Given that the diffraction limit for the human eye is approximately 60'', how far away from the surface of the Earth would one need to be to resolve objects with the same precision as the theorized gravitational telescope?**

\begin{align}
\alpha &= \frac{s}{r}
\end{align}

Solve for $r$:

\begin{align}
r &= \frac{s}{\alpha}
\end{align}

Let $s=17m$, and let $\alpha=60''$. Solve for $\alpha$ in radians:

\begin{align}
\frac{60''}{1}\cdot \frac{1'}{60''}\cdot \frac{1deg}{60'}\cdot\frac{2\pi rad}{360 deg} = 2.9\times 10^{-4}rad
\end{align}

Solve for $r$:

\begin{align}
r &= \frac{17}{2.9\times 10^{-4}}\\
&= 58620m\\
&= 58.62 km
\end{align}
